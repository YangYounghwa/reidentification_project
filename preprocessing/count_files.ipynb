```python
import os

def count_files_in_subfolders(folder1):
    total_sum = 0
    for folder2_name in os.listdir(folder1):
        folder2_path = os.path.join(folder1, folder2_name)

        if os.path.isdir(folder2_path):
            num_files = sum(
                len(files) for _, _, files in os.walk(folder2_path)
            )
            total_sum += num_files
            print(f"{folder2_name}: {num_files} files")
    print(total_sum)


```

```python
folder1 = r"..\data\dataset1"  
count_files_in_subfolders(folder1)
```
